```julia
using Revise
using DiffusionModels
using Images
using TestImages

includet("plotting.jl")
```

```julia
img = testimage("lighthouse")
img = imresize(img, (64, 64))
```

```julia
img_array = channelview(img);
img_array = permutedims(img_array, (2,3,1));
img_array = @. img_array * 2 - 1;
```

```julia
schedule = CosineSchedule()
# score_fn(x,p,t) = (x - img_array / std)
diffusion = VPDiffusion(schedule)
```

```julia
img_out = plot_diffusion_marginal(diffusion, img_array)
imresize(img_out, ratio=2)
```

```julia
img_out = plot_diffusion_forward_sde_transition(diffusion, img_array)
imresize(img_out, ratio=2)
```
