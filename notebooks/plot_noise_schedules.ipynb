```julia
using Revise
using DiffusionModels
using Plots
```

```julia
function plot_noise_schedule(s::NoiseSchedule)
    x = range(0, 1, length=1000)
    y = s.(x)
    p1 = plot(x, y, title="γ(t)")
    p2 = plot(x, sqrt.(y), title="√γ(t)")
    plot(p1, p2, layout=(1,2), legend=false)#, title=string(string(s)))
end
```

```julia
plot_noise_schedule(CosineSchedule())
```

```julia
plot_noise_schedule(SigmoidSchedule())
```

```julia
plot_noise_schedule(LinearSchedule())
```

```julia
plot_noise_schedule(ChenLinearSchedule())
```
